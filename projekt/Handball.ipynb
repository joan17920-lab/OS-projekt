```python
df = pd.merge(athlete_events_df, noc_regions_df, how='inner', on='NOC')

# Plot the age distribution by sex histogram
def plot_age_distribution_by_sex(sport_df, sport_name):
    sns.histplot(data=sport_df.dropna(subset=["Age"]), x="Age", hue="Sex", bins=20)
    plt.title(f'Age Distribution in {sport_name} By Sex')
    plt.xlabel('Age')
    plt.ylabel('Count')
    plt.show()

# Plot the age distribution histogram
def plot_age_distribution(sport_df, sport_name):
    sns.histplot(sport_df['Age'].dropna(), bins=20)
    plt.title(f'Age Distribution in {sport_name}')
    plt.xlabel('Age')
    plt.ylabel('Count')
    plt.show()

# Plot the medal distribution by sex histogram
def plot_distribution_medals_sex(sport_df, sport_name):
    sns.histplot(data=sport_df[sport_df["NOC"] == "AUT"].dropna(subset=["Medal"]).drop_duplicates(subset=["Event", "Year"]), x="Medal", hue="Sex", bins=20)
    plt.title(f'Medal Distribution in {sport_name} By Sex')
    plt.xlabel('Medal')
    plt.ylabel('Count')
    plt.show()

# Plot age distribution by gender
def plot_age_by_gender(sport_df, sport_name):
    sns.boxplot(x='Sex', y='Age', data=sport_df)
    plt.title(f'Age Distribution in {sport_name} by Gender')
    plt.xlabel('Gender')
    plt.ylabel('Age')
    plt.show()

def plot_best_result_age(sport_df, sport_name):
    
    pass


# Plot number of events per year
def plot_events_by_year(sport_df, sport_name):
    sns.countplot(x='Year', data=sport_df.drop_duplicates(subset=["Event","Year"]))
    plt.title(f'Number of {sport_name} Events by Year')
    plt.xlabel('Year')
    plt.ylabel('Number of Events')
    plt.xticks(rotation=45)
    plt.show()

# Run full sport analysis
def analyze_sport(merged_df, sport='Handball'):
    sport_df = merged_df[merged_df['Sport'] == sport].copy()
 
    plot_age_distribution(sport_df, sport)
    plot_age_distribution_by_sex(sport_df, sport)
    plot_distribution_medals_sex(sport_df, sport)
    plot_age_by_gender(sport_df, sport)
    plot_events_by_year(sport_df, sport)

analyze_sport(df, sport="Handball")
```
